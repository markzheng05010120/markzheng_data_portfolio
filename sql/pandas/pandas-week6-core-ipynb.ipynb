```python
import pandas as pd

df = pd.DataFrame({
    'country':['US','US','CN','CN'],
    'qty':[1,3,2,4],
    'price':[10,12,8,9]
})

display(df.head(), df.info(), df.describe())


df[['country','qty']].head(3)                     
high_qty = df[df['qty'] >= 2]                    
mid_price = df[df['price'].between(9,12)]        
high_qty.shape, mid_price.shape


df['amount'] = df['qty'] * df['price']
agg = (df.groupby('country')
         .agg(total_amount=('amount','sum'),
              orders=('country','count'))
         .reset_index()
         .sort_values('total_amount', ascending=False))
display(agg)


left = pd.DataFrame({'customer':['A','B','C'], 'country':['US','CN','US']})
right = pd.DataFrame({'country':['US','CN'], 'tax_rate':[0.08, 0.05]})
merged = left.merge(right, on='country', how='left')
display(merged)
```
